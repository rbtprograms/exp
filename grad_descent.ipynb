```python
#lets do a simple gradient descent
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_vals = 50
true_w = .5

x_vals = np.sort(5*(np.random.random(num_vals) - .05))
y_vals = true_w*(x_vals+x_vals**2) + np.random.normal(0,.1,num_vals)
y_vals[int(num_vals*.1):]+=3

plt.figure(figsize=(7,7))
plt.scatter(x_vals,y_vals)
plt.xlabel('x', fontsize=20)
plt.ylabel('y',fontsize=20)
plt.title('Data', fontsize=25)
```
