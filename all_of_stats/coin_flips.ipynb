```python
#suppose a coin has probability p of falling heads up. if we flip the coin many times, we would
#expect the proportion of heads to be near p. we will make this formal later. take p=.5 and
#n=1000 and simulate n coin flips. Plot the proportion of heads as a function of n. repeat for p=.3, p=.03

import random
import matplotlib.pyplot as plt 

class Coin:
    heads_per_trial = []
    def __init__(self, trials):
        heads = 0
        for i in range(1, trials):
            random_int = random.randint(0,1)
            if random_int == 0:
                heads += 1
            self.heads_per_trial.append(heads/i)
            
trials = 1000
res = Coin(trials)

y = res.heads_per_trial
x = range(1,trials) 
    
# plotting the points  
plt.plot(x, y) 
plt.axhline(y = 0.5, color = 'r', linestyle = '-')
    
plt.xlabel('trials') 
plt.ylabel('proportion of heads') 
    
# giving a title to my graph 
plt.title('p = .5') 
    
# function to show the plot 
plt.show() 
```
